```python
### CORRELOGRAM ###
## USED RASTER FILE CONTAINING 5 LIQUEFACTION SEVERITY DATA ##

import os
import rasterio
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Base directory containing the raster files
base_dir = r'D:\ARC GIS\KTM VALLEY\WGS_1984_UTM_ZONE_45N\METHOD 2\Classify_v2.2\Raster Classify (Natural Breaks)'

# Function to read raster data and flatten it
def read_raster(file_path):
    with rasterio.open(file_path) as src:
        return src.read(1).flatten()

# List of raster files
raster_files = ['RFM_LSM.TIF', 'SVM_LSM.TIF', 'ANN_LSM.TIF', 'XGB_LSM.TIF','DNN_LSM.TIF']

# Read and flatten the raster data
data = [read_raster(os.path.join(base_dir, file)) for file in raster_files]

# # Convert to a pandas DataFrame
# df = pd.DataFrame(data).T
# df.columns = ['RFM', 'SVM', 'ANN', 'XGB', 'DNN']

# # Calculate the correlation matrix
# correlation_matrix = df.corr()

# # Plot the correlogram
# plt.figure(figsize=(10, 8))
# sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=0, vmax=1)
# plt.title('Correlogram of Raster Data')
# plt.show()

# Convert to a pandas DataFrame
df = pd.DataFrame(data).T
df.columns = ['RFM', 'SVM', 'ANN', 'XGB', 'DNN']

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Plot the correlogram
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".5f", cmap='coolwarm', vmin=0.99, vmax=1, annot_kws={'size': 12})
plt.title('Correlogram of Raster Data')
plt.show()
```
