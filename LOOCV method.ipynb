```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['NDTW', 'Sedi_thick', 'Predom', 'Vs30', 'PGA', 'Geo_Form'] 
RFM = [0.838, 0.803, 0.798, 0.858, 0.812, 0.825]
SVM = [0.818, 0.805, 0.790, 0.828, 0.803, 0.815]
ANN = [0.838, 0.767, 0.778, 0.801, 0.814, 0.789]
XGB = [0.814, 0.778, 0.692, 0.761, 0.776, 0.789]
DNN = [0.796, 0.705, 0.744, 0.757, 0.805, 0.792]
baseline = [0.860, 0.838, 0.814, 0.787, 0.792]

# Create a figure and axes for each model
fig, axes = plt.subplots(5, 1, figsize=(8, 23))  # Create a figure with 5 subplots

# Set the y-axis limits to be the same for all subplots
ylim = (0.6, 0.9)

# Set the font size for the plot
plt.rcParams['font.size'] = 14

# Create bar plots for each model using seaborn
sns.barplot(x=labels, y=RFM, ax=axes[0])
for i, val in enumerate(RFM):
    axes[0].text(i, val, f"{val:.3f}", ha="center", va="bottom", fontsize=14)
axes[0].axhline(y=baseline[0], color='r', linestyle='--', label='All parameters used')
axes[0].set_title('RFM', fontsize=16)
axes[0].set_xlabel('Parameters removed', fontsize=14)
axes[0].set_ylabel('Cohen Kappa Score', fontsize=14)
axes[0].legend(loc='upper right', bbox_to_anchor=(1.0, 1), fontsize=12)
axes[0].set_ylim(ylim)

sns.barplot(x=labels, y=SVM, ax=axes[1])
for i, val in enumerate(SVM):
    axes[1].text(i, val, f"{val:.3f}", ha="center", va="bottom", fontsize=14)
axes[1].axhline(y=baseline[1], color='r', linestyle='--', label='All parameters used')
axes[1].set_title('SVM', fontsize=16)
axes[1].set_xlabel('Parameters removed', fontsize=14)
axes[1].set_ylabel('Cohen Kappa Score', fontsize=14)
axes[1].legend(loc='upper right', bbox_to_anchor=(1.0, 1), fontsize=12)
axes[1].set_ylim(ylim)

sns.barplot(x=labels, y=ANN, ax=axes[2])
for i, val in enumerate(ANN):
    axes[2].text(i, val, f"{val:.3f}", ha="center", va="bottom", fontsize=14)
axes[2].axhline(y=baseline[2], color='r', linestyle='--', label='All parameters used')
axes[2].set_title('ANN', fontsize=16)
axes[2].set_xlabel('Parameters removed', fontsize=14)
axes[2].set_ylabel('Cohen Kappa Score', fontsize=14)
axes[2].legend(loc='upper right', bbox_to_anchor=(1.0, 1), fontsize=12)
axes[2].set_ylim(ylim)

sns.barplot(x=labels, y=XGB, ax=axes[3])  # Plot XGB on the 4th subplot
for i, val in enumerate(XGB):
    axes[3].text(i, val, f"{val:.3f}", ha="center", va="bottom", fontsize=14)
axes[3].axhline(y=baseline[3], color='r', linestyle='--', label='All parameters used')
axes[3].set_title('XGB', fontsize=16)
axes[3].set_xlabel('Parameters removed', fontsize=14)
axes[3].set_ylabel('Cohen Kappa Score', fontsize=14)
axes[3].legend(loc='upper right', bbox_to_anchor=(1.0, 1), fontsize=12)
axes[3].set_ylim(ylim)

sns.barplot(x=labels, y=DNN, ax=axes[4])  # Plot DNN on the 5th subplot
for i, val in enumerate(DNN):
    axes[4].text(i, val, f"{val:.3f}", ha="center", va="bottom", fontsize=14)
axes[4].axhline(y=baseline[4], color='r', linestyle='--', label='All parameters used')
axes[4].set_title('DNN', fontsize=16)
axes[4].set_xlabel('Parameters removed', fontsize=14)
axes[4].set_ylabel('Cohen Kappa Score', fontsize=14)
axes[4].legend(loc='upper right', bbox_to_anchor=(1.0, 1), fontsize=12)
axes[4].set_ylim(ylim)

# Show the plots
plt.tight_layout()
plt.savefig('LOOCV.png')
plt.show()
```
